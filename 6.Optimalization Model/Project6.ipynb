```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
```

```python
car_df.info()
```

```python
car_df.head()
```

```python
categorical_column = car_df[['Make', 'Model',	'Type',	'Origin',	'DriveTrain']]
numerical_column = car_df[['MSRP',	'EngineSize',	'Cylinders',	'Horsepower',	'MPG_City',	'MPG_Highway',	'Weight',	'Wheelbase'	,'Length']]
```

```python
encoder = OneHotEncoder()
categorical_column = encoder.fit_transform(categorical_column).toarray()
```

```python
scalar = MinMaxScaler()
numerical_column = scalar.fit_transform(numerical_column)
```

```python
categorical_column = pd.DataFrame(categorical_column)
categorical_column
```

```python
numerical_column = pd.DataFrame(numerical_column)
numerical_column
```

```python
X_df = pd.concat([categorical_column,numerical_column], axis = 1)
X_df
```

```python
X = X_df.iloc[:, :-1].values
y = X_df.iloc[:, -1:].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
!pip install xgboost
```

```python
import xgboost as xgb
model = xgb.XGBRegressor(objective ='reg:squarederror', learning_rate = 1, max_depth = 20, n_estimators = 500)
model.fit(X_train, y_train)
```

```python
result = model.score(X_test, y_test)
print("R2 : {}".format(result))
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from math import sqrt
k = X_test.shape[1]
n = len(X_test)
RMSE = float(format(np.sqrt(mean_squared_error(y_test, y_predict)),'.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)

print('RMSE =',RMSE, '\nMSE =',MSE, '\nMAE =',MAE, '\nR2 =', r2, '\nAdjusted R2 =', adj_r2)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_gridsearch = {
    'max_depth': [3, 5, 7],
    'learning_rate': [0.1, 0.01, 0.001],
    'n_estimators': [100, 200, 300],
    'colsample_bytree': [0.6, 0.8, 1.0]
}
```

```python
grid_search = GridSearchCV(model, param_grid_gridsearch, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(X, y)
print("GridSearch Best Params:", grid_search.best_params_)
print("GridSearch Best RMSE:", np.sqrt(-grid_search.best_score_))
print("GridSearch Best R2:", grid_search.best_estimator_.score(X, y))
```

```python
param_dist_randomsearch = {
    'max_depth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.1, 0.5, 1],
    'n_estimators': [100, 200, 300, 400],
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0]
}
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV, cross_val_score
from skopt import BayesSearchCV
from skopt.space import Real, Integer, Categorical
```

```python
random_search = RandomizedSearchCV(model, param_dist_randomsearch, n_iter=100, scoring='neg_mean_squared_error', cv=5)
random_search.fit(X, y)
print("RandomSearch Best Params:", random_search.best_params_)
print("RandomSearch Best RMSE:", np.sqrt(-random_search.best_score_))
print("RandomSearch Best R2:", random_search.best_estimator_.score(X, y))
```

```python
param_space_bayesian = {
    'max_depth': Integer(3, 7),
    'learning_rate': Real(0.01, 1, prior='log-uniform'),
    'n_estimators': Integer(100, 300),
}
```

```python
bayesian_search = BayesSearchCV(model, param_space_bayesian, n_iter=100, scoring='neg_mean_squared_error', cv=5, verbose=5)
bayesian_search.fit(X, y)
```

```python
print("Bayesian Optimization Best Params:", bayesian_search.best_params_)
print("Bayesian Optimization Best RMSE:", np.sqrt(-bayesian_search.best_score_))
print("Bayesian Optimization Best R2:", bayesian_search.best_estimator_.score(X, y))
```
